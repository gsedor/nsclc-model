```python
df = pd.DataFrame(index = ['Coronavirus', 'Influenza A','Staphylococcus aureus','Mycoplasma spp.','Aspergillus flavus'],columns=['Integral Dose '])


df['Integral Dose'] = np.array([3200,1900,6600,8400,34900])

z=np.array([3200,1900,6600,8400,34900])
labels=['Coronavirus', 'Influenza A','Staphylococcus aureus','Mycoplasma spp.','Aspergillus flavus']
# [np.arange(50,251,1) *i for i in df['Integral Dose']
 
fig,ax= plt.subplots(figsize=(8,5))

flatui = ["#9b59b6", "#3498db", "#a3a3a3", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(sns.color_palette(flatui))

# for i,data in enumerate([np.arange(50,251,1) *i for i in df['Integral Dose']][::-1]):
#     ax.plot(np.arange(50,251,1),1/data/(1000*60),label=df.index[len(df.index)-1-i],lw=4)

for i in np.arange(4,-1,-1):
    x=np.arange(50,251,1)
    data = z[i]
    
    ax.plot(x,data*1000/(x*60000),label=labels[i],lw=3.5)
    

ax.legend(edgecolor='k',fontsize=11.5,loc=1,framealpha=.9)



# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)
#plt.yscale('log')
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(color=(.5,.5,.5), linestyle='-', linewidth=.75)
ax.set_ylabel('Time to 90% Sterilization (min.)',fontsize=16)
ax.set_xlabel(r'UV Intensity $(\mu W\,cm^{-2})$',fontsize=16)
plt.tight_layout()

ax.set_xlim(50,250)

ax.set_ylim(-1,12)

plt.savefig('/Users/geoffreysedor/Downloads/' + 'BSC_organisms2.png', dpi = 300)
```
